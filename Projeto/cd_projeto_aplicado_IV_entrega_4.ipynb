Cruzamento entre os Datasets e Transformação de Dados

Os cruzamentos foram realizados de forma gradativa, para assegurar que não ocorresse duplicação de dados.

```python
# Cruzamento gradativo
customers_location = dfs['customers'].merge(dfs['geolocation'], how='inner', left_on='customer_zip_code_prefix', right_on='geolocation_zip_code_prefix').drop_duplicates('customer_id', keep='first')
cusloc_order = customers_location.merge(dfs['orders'], how='inner', on='customer_id')
cuslocord_item = cusloc_order.merge(dfs['items'], how='inner', on='order_id')
cuslocordite_prod = cuslocord_item.merge(dfs['products'], how='inner', on='product_id')
cuslocordite_rev= cuslocordite_prod.merge(dfs['review'], how='left', on='order_id')

# Selecionando as colunas de interesse
final = cuslocordite_rev[['customer_id', 'customer_unique_id', 'customer_zip_code_prefix',
       'customer_city', 'customer_state',
       'geolocation_lat', 'geolocation_lng','order_id', 'order_status',
       'order_purchase_timestamp', 'order_approved_at',
       'order_delivered_carrier_date', 'order_delivered_customer_date',
       'order_estimated_delivery_date', 'order_item_id', 'product_id',
       'seller_id', 'shipping_limit_date', 'price', 'freight_value',
       'product_category_name', 'product_photos_qty',
       'review_id', 'review_score', 'review_comment_title',
       'review_comment_message', 'review_creation_date',
       'review_answer_timestamp']]
```

```python
# Garantindo que todas as colunas sejam datetime
datas = [
    'order_purchase_timestamp',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date',
    'shipping_limit_date',
    'review_creation_date',
    'review_answer_timestamp'
]

for data in datas:
    final[data] = pd.to_datetime(final[data], errors='coerce')  # Converter e tratar erros

# Criando coluna de tempo de entrega
final['delivery_time'] = (final['order_delivered_customer_date'] - final['order_purchase_timestamp']).dt.days

# Seleção do período de interesse
final = final[(final['order_purchase_timestamp'].dt.year > 2016)
              &
              (final['order_purchase_timestamp'] < pd.to_datetime('20180901'))
             ]
final = final.reset_index(drop=True)
final.info()
```

# **Modelo**

O código importa bibliotecas para manipulação de dados e operações numéricas com pandas e numpy, para visualização com matplotlib, para pré‑processamento e avaliação com scikit‑learn e para construção e treino de rede LSTM com TensorFlow Keras. Em seguida selecionam‑se os registros referentes a 2017 e 2018 e agregam‑se os preços por par mês e ano.Os meses são reindexados de 1 a 12 para garantir alinhamento temporal mesmo quando houver lacunas nos dados.

O vetor resultante é convertido para tipo float e reshapeado quando necessário para compatibilidade com o MinMaxScaler.

O MinMaxScaler é ajustado aos dados de treino e normaliza a série para a faixa entre zero e um. O objeto scaler é preservado para permitir inverter a escala das previsões posteriormente.

A preparação para o LSTM inclui a criação de janelas deslizantes por meio de uma função que gera vetores X de comprimento seq_length e alvos y correspondentes ao passo seguinte. Com seq_length igual a três realiza‑se o reshape de X para o formato tridimensional exigido por camadas recorrentes.

O modelo é um Sequential simples composto por uma camada LSTM com 64 unidades seguida de uma camada densa de saída única. O modelo é compilado com otimizador Adam e perda MSE e treinado por 200 épocas com batch size igual a oito.

A predição para os meses nove a doze de 2018 é feita de forma recursiva extraindo‑se a última janela normalizada do treino e gerando previsões passo a passo que são realimentadas como entrada para a iteração seguinte. As previsões normalizadas são então invertidas para a escala original usando o scaler e incorporadas à série de 2018 substituindo os meses previstos.

Para avaliação define‑se um limiar de binarização como a média combinada dos valores reais disponíveis em 2017 e em 2018 transformando séries contínuas em classes acima ou abaixo do limiar para cálculo de métricas de F1 quando aplicável. As métricas calculadas sobre as janelas usadas no treino mostraram R2 igual a 1.000 RMSE igual a 2.16 e MAE igual a 2,05 e F1 binarizado igual N/A. Esses resultados indicam ajuste aparente do modelo nos dados de treino mas também erros absolutos elevados e risco de sobreajuste.

Em termos práticos recomenda‑se estender e robustecer a validação temporal usando walk‑forward e janelas expansivas revisar o pré‑processamento e a interpolação testar estratégias de regularização ou simplificação do modelo e considerar alternativas à predição recursiva antes de confiar nas previsões para suporte à decisão. O gráfico final compara a soma mensal dos preços entre 2017 e 2018 mostrando dados reais até agosto de 2018 e previsões para setembro a dezembro.
```python
# Célula 1: Preparação, treino e previsões (execute antes das células de plot)
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import RobustScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, f1_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, RepeatVector, TimeDistributed
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import Huber
import warnings
warnings.filterwarnings("ignore")
np.random.seed(42)

# --- Assumptions: 'final' existe com order_purchase_timestamp e price ---
final['order_purchase_timestamp'] = pd.to_datetime(final['order_purchase_timestamp'])
final['year'] = final['order_purchase_timestamp'].dt.year
final['month'] = final['order_purchase_timestamp'].dt.month

monthly_prices = final[final['year'].isin([2017, 2018])].groupby(['month','year'])['price'].sum().unstack()
monthly_prices = monthly_prices.reindex(index=range(1,13))
monthly_prices = monthly_prices.apply(lambda col: col.interpolate(limit_direction='both'))

# Série de treino: 2017 completo + 2018 até agosto
series_2017 = monthly_prices[2017].values.astype(float)
series_2018_up_to_aug = monthly_prices[2018].iloc[:8].values.astype(float)
series = np.concatenate([series_2017, series_2018_up_to_aug])

if np.isnan(series).any():
    series = pd.Series(series).interpolate(limit_direction='both').values

# Winsorize simples (1%/99%) e log1p
low_p, high_p = np.nanpercentile(series, [1, 99])
series_clipped = np.clip(series, low_p, high_p)
series_log = np.log1p(series_clipped).astype(float)

# Features: rolling + sazonalidade
months_seq = np.concatenate([np.arange(1,13), np.arange(1,13)])[:len(series)]
sin_m = np.sin(2 * np.pi * months_seq / 12).reshape(-1,1)
cos_m = np.cos(2 * np.pi * months_seq / 12).reshape(-1,1)
roll3 = pd.Series(series_clipped).rolling(window=3, min_periods=1).mean().values.reshape(-1,1)
roll6 = pd.Series(series_clipped).rolling(window=6, min_periods=1).mean().values.reshape(-1,1)
value_log = series_log.reshape(-1,1)
features = np.hstack([value_log, roll3, roll6, sin_m, cos_m])

# Escalonamento robusto
scaler_X = RobustScaler()
X_scaled = scaler_X.fit_transform(features)

# Sequências e targets multi-step (seq_length -> horizon)
seq_length = 6
horizon = 4
X_seq = []
Y_seq = []
N = len(X_scaled)
for i in range(N - seq_length - horizon + 1):
    X_seq.append(X_scaled[i:i+seq_length])
    y_window = series_log[i+seq_length : i+seq_length+horizon].reshape(horizon, 1)
    Y_seq.append(y_window)

X_seq = np.array(X_seq)
Y_seq = np.array(Y_seq)

if X_seq.shape[0] == 0:
    raise ValueError("Sequências insuficientes: reduza seq_length ou horizon, ou verifique os dados.")

# Escalonar targets (RobustScaler)
scaler_y = RobustScaler()
Y_flat = Y_seq.reshape(-1, 1)
Y_flat_scaled = scaler_y.fit_transform(Y_flat)
Y_seq_scaled = Y_flat_scaled.reshape(Y_seq.shape)

# Split treino/val (últimas K janelas para validação)
K = 3
if X_seq.shape[0] > K:
    X_train, X_val = X_seq[:-K], X_seq[-K:]
    Y_train, Y_val = Y_seq_scaled[:-K], Y_seq_scaled[-K:]
    Y_train_true_log = Y_seq[:-K].reshape(-1,1)
    Y_val_true_log = Y_seq[-K:].reshape(-1,1)
else:
    X_train, X_val = X_seq, X_seq
    Y_train, Y_val = Y_seq_scaled, Y_seq_scaled
    Y_train_true_log = Y_seq.reshape(-1,1)
    Y_val_true_log = Y_seq.reshape(-1,1)

# Modelo encoder-decoder LSTM (saída multi-step)
n_features = X_seq.shape[2]
model = Sequential([
    LSTM(128, return_sequences=False, input_shape=(seq_length, n_features)),
    Dropout(0.2),
    RepeatVector(horizon),
    LSTM(64, return_sequences=True),
    Dropout(0.2),
    TimeDistributed(Dense(1))
])
opt = Adam(learning_rate=1e-3, clipnorm=1.0)
model.compile(optimizer=opt, loss=Huber(delta=1.0))

callbacks = [
    EarlyStopping(monitor='val_loss', patience=40, restore_best_weights=True, verbose=0),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=12, min_lr=1e-6, verbose=0)
]

history = model.fit(X_train, Y_train, validation_data=(X_val, Y_val),
                    epochs=400, batch_size=8, callbacks=callbacks, shuffle=False, verbose=1)

# Funções utilitárias
def inv_transform_preds(pred_scaled, scaler_y):
    flat = pred_scaled.reshape(-1,1)
    inv_flat = scaler_y.inverse_transform(flat)
    inv_log = inv_flat.flatten()
    inv_orig = np.expm1(inv_log)
    return inv_orig.reshape(pred_scaled.shape[0], pred_scaled.shape[1])

def compute_metrics(y_true_orig, y_pred_orig, threshold):
    rmse = np.sqrt(mean_squared_error(y_true_orig, y_pred_orig))
    mae = mean_absolute_error(y_true_orig, y_pred_orig)
    r2 = r2_score(y_true_orig, y_pred_orig) if len(y_true_orig) > 1 else np.nan
    y_true_bin = (y_true_orig > threshold).astype(int)
    y_pred_bin = (y_pred_orig > threshold).astype(int)
    f1 = f1_score(y_true_bin, y_pred_bin) if len(np.unique(y_true_bin)) > 1 else np.nan
    return rmse, mae, r2, f1

# Avaliação treino/val (janelas)
pred_train_scaled = model.predict(X_train, verbose=0)
pred_train = inv_transform_preds(pred_train_scaled, scaler_y)
y_train_true = np.expm1(Y_train_true_log).flatten()
y_train_pred = pred_train.flatten()
rmse_train, mae_train, r2_train, f1_train = compute_metrics(y_train_true, y_train_pred,
                                                            np.nanmean(np.concatenate([monthly_prices[2017].dropna().values, monthly_prices[2018].dropna().values])))

pred_val_scaled = model.predict(X_val, verbose=0)
pred_val = inv_transform_preds(pred_val_scaled, scaler_y)
y_val_true = np.expm1(Y_val_true_log).flatten()
y_val_pred = pred_val.flatten()
rmse_val, mae_val, r2_val, f1_val = compute_metrics(y_val_true, y_val_pred,
                                                    np.nanmean(np.concatenate([monthly_prices[2017].dropna().values, monthly_prices[2018].dropna().values])))

print("=== Avaliação no Treino (sobre janelas usadas para treinar) ===")
print(f"RMSE (treino): {rmse_train:.2f}")
print(f"MAE  (treino): {mae_train:.2f}")
print(f"R2   (treino): {r2_train:.4f}")
print(f"F1 binarizado (treino): {f1_train if not np.isnan(f1_train) else 'N/A (classe única)'}\n")

print("=== Validação Retida (últimas K janelas) ===")
print(f"RMSE (val): {rmse_val:.2f}")
print(f"MAE  (val): {mae_val:.2f}")
print(f"R2   (val): {r2_val:.4f}")
print(f"F1 binarizado (val): {f1_val if not np.isnan(f1_val) else 'N/A (classe única)'}\n")

# Previsão direta para meses 9..12 de 2018
last_window = X_scaled[-seq_length:].reshape(1, seq_length, n_features)
pred_future_scaled = model.predict(last_window, verbose=0)
pred_future = inv_transform_preds(pred_future_scaled, scaler_y).flatten()

full_2018 = monthly_prices[2018].copy()
for i, m in enumerate(range(9, 13)):
    full_2018.loc[m] = pred_future[i]

pred_months = [9,10,11,12]
pred_df = pd.DataFrame({'month': pred_months, 'predicted_price': pred_future})
print("\nPrevisões para 2018 (meses 9-12):")
print(pred_df)
```
