Cruzamento entre os Datasets e Transformação de Dados

Os cruzamentos foram realizados de forma gradativa, para assegurar que não ocorresse duplicação de dados.

```python
# Cruzamento gradativo
customers_location = dfs['customers'].merge(dfs['geolocation'], how='inner', left_on='customer_zip_code_prefix', right_on='geolocation_zip_code_prefix').drop_duplicates('customer_id', keep='first')
cusloc_order = customers_location.merge(dfs['orders'], how='inner', on='customer_id')
cuslocord_item = cusloc_order.merge(dfs['items'], how='inner', on='order_id')
cuslocordite_prod = cuslocord_item.merge(dfs['products'], how='inner', on='product_id')
cuslocordite_rev= cuslocordite_prod.merge(dfs['review'], how='left', on='order_id')

# Selecionando as colunas de interesse
final = cuslocordite_rev[['customer_id', 'customer_unique_id', 'customer_zip_code_prefix',
       'customer_city', 'customer_state',
       'geolocation_lat', 'geolocation_lng','order_id', 'order_status',
       'order_purchase_timestamp', 'order_approved_at',
       'order_delivered_carrier_date', 'order_delivered_customer_date',
       'order_estimated_delivery_date', 'order_item_id', 'product_id',
       'seller_id', 'shipping_limit_date', 'price', 'freight_value',
       'product_category_name', 'product_photos_qty',
       'review_id', 'review_score', 'review_comment_title',
       'review_comment_message', 'review_creation_date',
       'review_answer_timestamp']]
```

O código inicia com a importação das bibliotecas necessárias para manipulação de dados (pandas, numpy), visualização (matplotlib), pré-processamento e avaliação (scikit-learn), e construção de redes neurais LSTM (TensorFlow Keras). São geradas variáveis auxiliares como médias móveis de 3 e 6 meses, além de componentes sazonais baseados em funções seno e cosseno. Essas variáveis são combinadas em uma matriz multivariada de features e escaladas com RobustScaler.

A estrutura de entrada do modelo é organizada em janelas deslizantes com comprimento de 6 meses (seq_length = 6) e horizonte de previsão de 4 meses (horizon = 4). Os alvos (targets) também são transformados com log1p e escalados com RobustScaler. A divisão entre treino e validação é feita separando as três últimas janelas para validação, mantendo a ordem cronológica dos dados.

O modelo implementado é uma rede LSTM encoder-decoder com saída multistep. A arquitetura inclui uma camada LSTM com 128 unidades, seguida de Dropout, RepeatVector, uma segunda camada LSTM com 64 unidades, outro Dropout e uma camada TimeDistributed(Dense) para gerar múltiplas previsões. A função de perda utilizada é Huber, e o otimizador é Adam com taxa de aprendizado inicial de 0,001 e clipagem de gradiente.

Durante o treinamento, são utilizados os callbacks EarlyStopping e ReduceLROnPlateau para evitar overfitting e ajustar dinamicamente a taxa de aprendizado. O modelo é treinado por até 400 épocas com batch_size = 8, sem embaralhamento dos dados (shuffle=False) e com sementes fixadas para garantir reprodutibilidade. Após o treino, as previsões são revertidas da escala logarítmica para os valores originais.

```python
# Célula 1: Preparação, treino e previsões (execute antes das células de plot)
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import RobustScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, f1_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, RepeatVector, TimeDistributed
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import Huber
import warnings
warnings.filterwarnings("ignore")
np.random.seed(42)

# --- Assumptions: 'final' existe com order_purchase_timestamp e price ---
final['order_purchase_timestamp'] = pd.to_datetime(final['order_purchase_timestamp'])
final['year'] = final['order_purchase_timestamp'].dt.year
final['month'] = final['order_purchase_timestamp'].dt.month

monthly_prices = final[final['year'].isin([2017, 2018])].groupby(['month','year'])['price'].sum().unstack()
monthly_prices = monthly_prices.reindex(index=range(1,13))
monthly_prices = monthly_prices.apply(lambda col: col.interpolate(limit_direction='both'))

# Série de treino: 2017 completo + 2018 até maio
series_2017 = monthly_prices[2017].values.astype(float)
series_2018_up_to_may = monthly_prices[2018].iloc[:5].values.astype(float)  # meses 1..5
series = np.concatenate([series_2017, series_2018_up_to_may])

if np.isnan(series).any():
    series = pd.Series(series).interpolate(limit_direction='both').values

# Winsorize simples (1%/99%) e log1p
low_p, high_p = np.nanpercentile(series, [1, 99])
series_clipped = np.clip(series, low_p, high_p)
series_log = np.log1p(series_clipped).astype(float)

# Features: rolling + sazonalidade
months_seq = np.concatenate([np.arange(1,13), np.arange(1,13)])[:len(series)]
sin_m = np.sin(2 * np.pi * months_seq / 12).reshape(-1,1)
cos_m = np.cos(2 * np.pi * months_seq / 12).reshape(-1,1)
roll3 = pd.Series(series_clipped).rolling(window=3, min_periods=1).mean().values.reshape(-1,1)
roll6 = pd.Series(series_clipped).rolling(window=6, min_periods=1).mean().values.reshape(-1,1)
value_log = series_log.reshape(-1,1)
features = np.hstack([value_log, roll3, roll6, sin_m, cos_m])

# Escalonamento robusto
scaler_X = RobustScaler()
X_scaled = scaler_X.fit_transform(features)

# Sequências e targets multi-step (seq_length -> horizon)
seq_length = 6
horizon = 4   # previsão de jun, jul, ago, set
X_seq, Y_seq = [], []
N = len(X_scaled)
for i in range(N - seq_length - horizon + 1):
    X_seq.append(X_scaled[i:i+seq_length])
    y_window = series_log[i+seq_length : i+seq_length+horizon].reshape(horizon, 1)
    Y_seq.append(y_window)

X_seq = np.array(X_seq)
Y_seq = np.array(Y_seq)

# Escalonar targets (RobustScaler)
scaler_y = RobustScaler()
Y_flat = Y_seq.reshape(-1, 1)
Y_flat_scaled = scaler_y.fit_transform(Y_flat)
Y_seq_scaled = Y_flat_scaled.reshape(Y_seq.shape)

# Split treino/val (últimas K janelas para validação)
K = 3
X_train, X_val = X_seq[:-K], X_seq[-K:]
Y_train, Y_val = Y_seq_scaled[:-K], Y_seq_scaled[-K:]
Y_train_true_log = Y_seq[:-K].reshape(-1,1)
Y_val_true_log = Y_seq[-K:].reshape(-1,1)

# Modelo encoder-decoder LSTM (saída multi-step)
n_features = X_seq.shape[2]
model = Sequential([
    LSTM(128, return_sequences=False, input_shape=(seq_length, n_features)),
    Dropout(0.2),
    RepeatVector(horizon),
    LSTM(64, return_sequences=True),
    Dropout(0.2),
    TimeDistributed(Dense(1))
])
opt = Adam(learning_rate=1e-3, clipnorm=1.0)
model.compile(optimizer=opt, loss=Huber(delta=1.0))

callbacks = [
    EarlyStopping(monitor='val_loss', patience=40, restore_best_weights=True, verbose=0),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=12, min_lr=1e-6, verbose=0)
]

history = model.fit(X_train, Y_train, validation_data=(X_val, Y_val),
                    epochs=400, batch_size=8, callbacks=callbacks, shuffle=False, verbose=1)

# Funções utilitárias
def inv_transform_preds(pred_scaled, scaler_y):
    flat = pred_scaled.reshape(-1,1)
    inv_flat = scaler_y.inverse_transform(flat)
    inv_log = inv_flat.flatten()
    inv_orig = np.expm1(inv_log)
    return inv_orig.reshape(pred_scaled.shape[0], pred_scaled.shape[1])

def compute_metrics(y_true_orig, y_pred_orig, threshold):
    rmse = np.sqrt(mean_squared_error(y_true_orig, y_pred_orig))
    mae = mean_absolute_error(y_true_orig, y_pred_orig)
    r2 = r2_score(y_true_orig, y_pred_orig) if len(y_true_orig) > 1 else np.nan
    y_true_bin = (y_true_orig > threshold).astype(int)
    y_pred_bin = (y_pred_orig > threshold).astype(int)
    f1 = f1_score(y_true_bin, y_pred_bin) if len(np.unique(y_true_bin)) > 1 else np.nan
    return rmse, mae, r2, f1

# Avaliação treino/val (janelas)
pred_train_scaled = model.predict(X_train, verbose=0)
pred_train = inv_transform_preds(pred_train_scaled, scaler_y)
y_train_true = np.expm1(Y_train_true_log).flatten()
y_train_pred = pred_train.flatten()
rmse_train, mae_train, r2_train, f1_train = compute_metrics(
    y_train_true, y_train_pred,
    np.nanmean(np.concatenate([monthly_prices[2017].dropna().values,
                               monthly_prices[2018].dropna().values]))
)

pred_val_scaled = model.predict(X_val, verbose=0)
pred_val = inv_transform_preds(pred_val_scaled, scaler_y)
y_val_true = np.expm1(Y_val_true_log).flatten()
y_val_pred = pred_val.flatten()
rmse_val, mae_val, r2_val, f1_val = compute_metrics(
    y_val_true, y_val_pred,
    np.nanmean(np.concatenate([monthly_prices[2017].dropna().values,
                               monthly_prices[2018].dropna().values]))
)

# >>> BLOCO DE PRINTS DE AVALIAÇÃO <<<
print("=== Avaliação no Treino (sobre janelas usadas para treinar) ===")
print(f"RMSE (treino): {rmse_train:.2f}")
print(f"MAE  (treino): {mae_train:.2f}")
print(f"R2   (treino): {r2_train:.4f}")
print(f"F1 binarizado (treino): {f1_train if not np.isnan(f1_train) else 'N/A (classe única)'}\n")

print("=== Validação Retida (últimas K janelas) ===")
print(f"RMSE (val): {rmse_val:.2f}")
print(f"MAE  (val): {mae_val:.2f}")
print(f"R2   (val): {r2_val:.4f}")
print(f"F1 binarizado (val): {f1_val if not np.isnan(f1_val) else 'N/A (classe única)'}\n")

# Previsão direta para meses 6..9 de 2018
last_window = X_scaled[-seq_length:].reshape(1, seq_length, n_features)
pred_future_scaled = model.predict(last_window, verbose=0)
pred_future = inv_transform_preds(pred_future_scaled, scaler_y).flatten()

pred_months = list(range(6, 6+horizon))  # meses 6..9
pred_df = pd.DataFrame({
    'month': pred_months,
    'predicted_price': pred_future
})
print("\nPrevisões para 2018 (meses 6-9):")
print(pred_df)

# Comparação com realizado
real_values = monthly_prices[2018].loc[pred_months].values
comparison = pd.DataFrame({
    'month': pred_months,
    'real_price': real_values,
    'predicted_price': pred_future
})
print("\nComparação previsto vs realizado (jun-set 2018):")
print(comparison)
```

```python
# Célula 1: Preparação, treino e previsões com df_completo
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import RobustScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, f1_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, RepeatVector, TimeDistributed
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import Huber
import warnings
warnings.filterwarnings("ignore")
np.random.seed(42)

# Garantir datetime e extrair ano/mês
df_completo['order_purchase_timestamp'] = pd.to_datetime(df_completo['order_purchase_timestamp'])
df_completo['year'] = df_completo['order_purchase_timestamp'].dt.year
df_completo['month'] = df_completo['order_purchase_timestamp'].dt.month

# Agrupar por mês e ano
monthly_prices = df_completo[df_completo['year'].isin([2017, 2018])].groupby(['month','year'])['price'].sum().unstack()
monthly_prices = monthly_prices.reindex(index=range(1,13))
monthly_prices = monthly_prices.apply(lambda col: col.interpolate(limit_direction='both'))

# Série de treino: 2017 completo + 2018 até maio
series_2017 = monthly_prices[2017].values.astype(float)
series_2018_up_to_may = monthly_prices[2018].iloc[:5].values.astype(float)
series = np.concatenate([series_2017, series_2018_up_to_may])

if np.isnan(series).any():
    series = pd.Series(series).interpolate(limit_direction='both').values

# Winsorize e log1p
low_p, high_p = np.nanpercentile(series, [1, 99])
series_clipped = np.clip(series, low_p, high_p)
series_log = np.log1p(series_clipped).astype(float)

# Features: rolling + sazonalidade
months_seq = np.concatenate([np.arange(1,13), np.arange(1,13)])[:len(series)]
sin_m = np.sin(2 * np.pi * months_seq / 12).reshape(-1,1)
cos_m = np.cos(2 * np.pi * months_seq / 12).reshape(-1,1)
roll3 = pd.Series(series_clipped).rolling(window=3, min_periods=1).mean().values.reshape(-1,1)
roll6 = pd.Series(series_clipped).rolling(window=6, min_periods=1).mean().values.reshape(-1,1)
value_log = series_log.reshape(-1,1)
features = np.hstack([value_log, roll3, roll6, sin_m, cos_m])

# Escalonamento
scaler_X = RobustScaler()
X_scaled = scaler_X.fit_transform(features)

# Sequências e targets
seq_length = 6
horizon = 4
X_seq, Y_seq = [], []
N = len(X_scaled)
for i in range(N - seq_length - horizon + 1):
    X_seq.append(X_scaled[i:i+seq_length])
    y_window = series_log[i+seq_length : i+seq_length+horizon].reshape(horizon, 1)
    Y_seq.append(y_window)

X_seq = np.array(X_seq)
Y_seq = np.array(Y_seq)

# Escalonar targets
scaler_y = RobustScaler()
Y_flat = Y_seq.reshape(-1, 1)
Y_flat_scaled = scaler_y.fit_transform(Y_flat)
Y_seq_scaled = Y_flat_scaled.reshape(Y_seq.shape)

# Split treino/val
K = 3
X_train, X_val = X_seq[:-K], X_seq[-K:]
Y_train, Y_val = Y_seq_scaled[:-K], Y_seq_scaled[-K:]
Y_train_true_log = Y_seq[:-K].reshape(-1,1)
Y_val_true_log = Y_seq[-K:].reshape(-1,1)

# Modelo LSTM
n_features = X_seq.shape[2]
model = Sequential([
    LSTM(128, return_sequences=False, input_shape=(seq_length, n_features)),
    Dropout(0.2),
    RepeatVector(horizon),
    LSTM(64, return_sequences=True),
    Dropout(0.2),
    TimeDistributed(Dense(1))
])
opt = Adam(learning_rate=1e-3, clipnorm=1.0)
model.compile(optimizer=opt, loss=Huber(delta=1.0))

callbacks = [
    EarlyStopping(monitor='val_loss', patience=40, restore_best_weights=True, verbose=0),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=12, min_lr=1e-6, verbose=0)
]

history = model.fit(X_train, Y_train, validation_data=(X_val, Y_val),
                    epochs=400, batch_size=8, callbacks=callbacks, shuffle=False, verbose=1)

# Funções utilitárias
def inv_transform_preds(pred_scaled, scaler_y):
    flat = pred_scaled.reshape(-1,1)
    inv_flat = scaler_y.inverse_transform(flat)
    inv_log = inv_flat.flatten()
    inv_orig = np.expm1(inv_log)
    return inv_orig.reshape(pred_scaled.shape[0], pred_scaled.shape[1])

def compute_metrics(y_true_orig, y_pred_orig, threshold):
    rmse = np.sqrt(mean_squared_error(y_true_orig, y_pred_orig))
    mae = mean_absolute_error(y_true_orig, y_pred_orig)
    r2 = r2_score(y_true_orig, y_pred_orig) if len(y_true_orig) > 1 else np.nan
    y_true_bin = (y_true_orig > threshold).astype(int)
    y_pred_bin = (y_pred_orig > threshold).astype(int)
    f1 = f1_score(y_true_bin, y_pred_bin) if len(np.unique(y_true_bin)) > 1 else np.nan
    return rmse, mae, r2, f1

# Avaliação treino/val
pred_train_scaled = model.predict(X_train, verbose=0)
pred_train = inv_transform_preds(pred_train_scaled, scaler_y)
y_train_true = np.expm1(Y_train_true_log).flatten()
y_train_pred = pred_train.flatten()
rmse_train, mae_train, r2_train, f1_train = compute_metrics(
    y_train_true, y_train_pred,
    np.nanmean(np.concatenate([monthly_prices[2017].dropna().values,
                               monthly_prices[2018].dropna().values]))
)

pred_val_scaled = model.predict(X_val, verbose=0)
pred_val = inv_transform_preds(pred_val_scaled, scaler_y)
y_val_true = np.expm1(Y_val_true_log).flatten()
y_val_pred = pred_val.flatten()
rmse_val, mae_val, r2_val, f1_val = compute_metrics(
    y_val_true, y_val_pred,
    np.nanmean(np.concatenate([monthly_prices[2017].dropna().values,
                               monthly_prices[2018].dropna().values]))
)

# Prints de avaliação
print("=== Avaliação no Treino ===")
print(f"RMSE (treino): {rmse_train:.2f}")
print(f"MAE  (treino): {mae_train:.2f}")
print(f"R2   (treino): {r2_train:.4f}")
print(f"F1 binarizado (treino): {f1_train if not np.isnan(f1_train) else 'N/A'}\n")

print("=== Validação Retida ===")
print(f"RMSE (val): {rmse_val:.2f}")
print(f"MAE  (val): {mae_val:.2f}")
print(f"R2   (val): {r2_val:.4f}")
print(f"F1 binarizado (val): {f1_val if not np.isnan(f1_val) else 'N/A'}\n")

# Previsão para meses 6 a 9 de 2018
last_window = X_scaled[-seq_length:].reshape(1, seq_length, n_features)
pred_future_scaled = model.predict(last_window, verbose=0)
pred_future = inv_transform_preds(pred_future_scaled, scaler_y).flatten()

pred_months = list(range(6, 6+horizon))
pred_df = pd.DataFrame({
    'month': pred_months,
    'predicted_price': pred_future
})
print("\nPrevisões para 2018 (meses 6-9):")
print(pred_df)

# Comparação com realizado
real_values = monthly_prices[2018].loc[pred_months].values
comparison = pd.DataFrame({
    'month': pred_months,
    'real_price': real_values,
    'predicted_price': pred_future
})
print("\nComparação previsto vs realizado (jun-set 2018):")
print(comparison)
```

Esses resultados sugerem que, nesse intervalo, o modelo apresentou uma margem de acerto inferior ao esperado, possivelmente impactada pela alta variabilidade entre os meses. Para melhorar a precisão, seria recomendável ajustar o modelo e incorporar um volume maior de dados históricos do produto avaliado, o que permitiria capturar melhor os padrões sazonais e comportamentais.
```python
# @title
import numpy as np
import pandas as pd
import calendar

# Definir meses previstos (jun–ago)
pred_months = [6, 7, 8]

# Valores reais (do monthly_prices)
real_values = monthly_prices[2018].loc[pred_months].values.astype(float)

# Valores previstos (do modelo já treinado)
pred_values = pred_future[:len(pred_months)]

# Construir DataFrame comparativo
df_pred = pd.DataFrame({
    'month': pred_months,
    'month_name': [calendar.month_name[m] for m in pred_months],
    'real_price': real_values,
    'predicted_price': pred_values
})

# Diferença absoluta e variação percentual
df_pred['diff'] = df_pred['predicted_price'] - df_pred['real_price']
df_pred['pct_change'] = np.where(
    df_pred['real_price'] == 0,
    np.nan,
    df_pred['diff'] / df_pred['real_price'] * 100
)

# Totais
total_real = df_pred['real_price'].sum()
total_pred = df_pred['predicted_price'].sum()
total_diff = total_pred - total_real
total_pct = (total_diff / total_real * 100) if total_real != 0 else np.nan

tot_row = pd.DataFrame({
    'month': [0],
    'month_name': ['Total'],
    'real_price': [total_real],
    'predicted_price': [total_pred],
    'diff': [total_diff],
    'pct_change': [total_pct]
})

df_tot = pd.concat([df_pred, tot_row], ignore_index=True)

# Formatação para exibição
df_display = df_tot.copy()
df_display[['real_price', 'predicted_price', 'diff']] = df_display[['real_price', 'predicted_price', 'diff']].round(2)
df_display['pct_change'] = df_display['pct_change'].apply(lambda x: f"{x:.2f}%" if pd.notna(x) else 'N/A')

# Exibir tabela
print(df_display[['month', 'month_name', 'real_price', 'predicted_price', 'diff', 'pct_change']])
```
