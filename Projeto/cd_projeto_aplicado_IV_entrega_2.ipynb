# **Introdução**

O comércio eletrônico foi consolidado como um dos principais vetores de crescimento econômico na era digital, impulsionado pela expansão do acesso à internet, pela evolução das tecnologias de informação e comunicação e pela diversificação dos hábitos de consumo. Por meio da análise temporal, serão identificadas tendências, sazonalidades e ciclos de demanda que contribuem para o equilíbrio entre oferta e demanda, a otimização de estoques e a minimização de desperdícios.

Espera se que os resultados possibilitem a redução dos custos de estoque entre 10% e 20%, o aumento da taxa de atendimento ao cliente em até 15% e a promoção de práticas comerciais mais sustentáveis por meio da prevenção de superestoques e da obsolescência de produtos. # **Pipeline da Solução**

O pipeline de processamento de dados adotado no projeto segue uma sequência estruturada de etapas que asseguram qualidade, eficiência e escalabilidade. As fontes de dados incluíram arquivos CSV e bases relacionais contendo informações de consumidores, geolocalização, pedidos, itens, produtos, vendedores e avaliações. A ingestão e a integração desses dados foram realizadas principalmente com Python e pandas, enquanto ferramentas de orquestração como Apache NiFi e Talend foram consideradas quando necessárias para automação de fluxos maiores. O levantamento inicial teve como objetivo garantir cobertura suficiente do domínio do problema e permitir a agregação temporal adequada para modelagem.

No pré processamento foram executadas checagens coluna a coluna para identificação e correção de inconsistências, remoção de duplicados após joins e tratamento de valores nulos. Para séries temporais utilizou‑se interpolação bidirecional quando apropriado para preservar continuidade temporal. Campos de data foram convertidos para o tipo datetime e foi criada a variável tempo de entrega como a diferença entre data de entrega e data do pedido. Para o escalonamento aplicou‑se MinMaxScaler ajustado apenas no conjunto de treino concatenado antes do treino do modelo. Essas ações reduziram ruído e melhoraram a qualidade dos sinais mantidos para modelagem.

A análise exploratória teve papel orientador nas decisões de agregação e seleção de features. Estatísticas descritivas, identificação de outliers e visualizações como histogramas e boxplots foram usadas para entender distribuições por variável e por mês. Notebooks em Jupyter e bibliotecas como Matplotlib e Seaborn documentaram padrões de sazonalidade e comportamento espacial quando aplicável, e esse entendimento levou à decisão de agregar os dados em base mensal para as previsões alvo.

A seleção de características combinou análise de correlação e métodos de importância para reduzir dimensionalidade sem perder sinais relevantes para previsão mensal com LSTM. Essa etapa buscou melhorar desempenho do modelo e facilitar interpretação das features escolhidas. Além das variáveis temporais básicas foram incluídas features exógenas como indicador de promoções, preço médio e flags de feriado e fim de semana.

Na modelagem foi empregada uma rede LSTM sequencial treinada sobre janelas temporais extraídas da série agregada. O conjunto de treino foi composto pelo ano completo de 2017 mais parte de 2018 até agosto, com NaNs interpolados quando necessário. Criaram‑se janelas móveis de entrada com sequência de comprimento ajustada por validação e as entradas foram escalonadas conforme descrito anteriormente. A arquitetura considerou uma camada LSTM com unidades suficientes para capturar dependências de longo prazo, seguida por uma camada densa de saída. Recursos de regularização como dropout e callbacks para early stopping foram previstos para controlar overfitting.

O treinamento utilizou função de perda MSE e otimizador Adam. Hiperparâmetros iniciais de referência incluíram epochs em torno de 200 e batch size pequeno para séries curtas, ajustados conforme validação. A validação temporal preservou a ordem cronológica e foi realizada por retenção das últimas janelas e por rolling window quando se buscou avaliação mais robusta. Para previsões multi horizonte os valores previstos foram gerados de forma recursiva, alimentando a próxima janela, e as saídas foram desscalonadas com o scaler original antes da integração aos processos de decisão.

A avaliação do desempenho aplicou métricas como RMSE, MAE e R2 em treino, validação retida e nas previsões para os meses alvo. MAPE e intervalos de confiança foram considerados para iterações futuras quando adequado. Uma métrica binarizada baseada em threshold foi incluída como opção para cenários específicos e reportada como N/A quando a granularidade dos rótulos não permitiu sua aplicação.

Em suma o pipeline integra coleta, limpeza, transformação, exploração, seleção de features, modelagem e validação de modo encadeado e pragmático. Cada etapa foi implementada com foco em preservar continuidade temporal e em reduzir vazamento de informação, garantindo que as previsões mensais geradas pelo modelo LSTM possam ser incorporadas de forma confiável ao planejamento de estoque e às decisões logísticas.
