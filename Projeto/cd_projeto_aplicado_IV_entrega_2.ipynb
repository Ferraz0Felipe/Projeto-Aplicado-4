# **Introdução**

O comércio eletrônico tem se consolidado como um dos principais vetores de crescimento econômico na era digital, apoiado pela expansão do acesso à internet, pela evolução das tecnologias de informação e comunicação e pela diversificação dos hábitos de consumo. No Brasil, esse setor alcançou faturamento de R$ 61,9 bilhões em 2018, com crescimento médio anual de 16,5 % entre 2016 e 2018, mas ainda enfrenta desafios como ruptura de estoque, custos elevados de armazenagem e falta de previsibilidade da demanda.

Essas lacunas motivam a aplicação de modelos de séries temporais e técnicas de machine learning para extrair conhecimento de dados históricos, reduzir perdas estimadas em até 5 % do faturamento anual por falta de previsão adequada e apoiar tomadas de decisão mais informadas em logística e marketing. A análise temporal pode identificar tendências, sazonalidades e ciclos de demanda, contribuindo para equilibrar oferta e demanda, otimizar estoques e minimizar desperdícios.

O objetivo geral deste projeto é desenvolver um sistema de análise de séries temporais para estudar padrões de vendas no comércio eletrônico brasileiro, utilizando como base o Olist e-Commerce Dataset (2016–2018) disponível no Kaggle. A partir da identificação de comportamentos históricos de compra, busca-se gerar previsões acuradas de volume de pedidos e faturamento que subsidiem decisões estratégicas em plataformas de e-commerce.

Objetivos específicos:

analisar a evolução temporal dos volumes de pedidos e do faturamento em diferentes horizontes (diário, semanal e mensal) por produto e região

identificar padrões de tendência, sazonalidade e ciclos de demanda que impactam o desempenho de vendas

comparar o desempenho de modelos de séries temporais (ARIMA, ETS e Prophet) e de machine learning (LSTM e XGBoost)

propor diretrizes para aplicação dos resultados na gestão de estoque, planejamento logístico e campanhas de marketing digital

A justificativa reside no potencial de reduzir de 10 % a 20 % os custos de estoque, aumentar em até 15 % a taxa de atendimento ao cliente e promover práticas comerciais sustentáveis ao evitar superestoques e produtos obsoletos. Além disso, o projeto está alinhado aos Objetivos de Desenvolvimento Sustentável da ONU, especialmente nos eixos de crescimento econômico, inovação tecnológica e consumo responsável.



# **Referencial Teórico**
A solução proposta baseia-se nos princípios do pensamento computacional, uma abordagem que permite resolver problemas complexos de forma lógica, estruturada e eficiente. Os principais conceitos envolvidos incluem:

Decomposição: divisão de um problema em partes menores e mais gerenciáveis.

Reconhecimento de padrões: identificação de semelhanças entre problemas ou dados, facilitando a reutilização de soluções.

Abstração: foco nos elementos essenciais do problema, ignorando detalhes irrelevantes.

Algoritmos: criação de instruções passo a passo para resolver o problema de forma automatizada (WING, 2006; SOMAI, 2023).

Esses pilares são amplamente discutidos por Wing (2006), que popularizou o termo “Computational Thinking” como uma habilidade essencial para o século XXI. Trabalhos como o de Silva (2020) também reforçam a importância dessa abordagem na educação e na resolução de problemas em diversas áreas, destacando sua aplicabilidade além da programação.

Alternativas de solução para problemas semelhantes incluem:

Inteligência Artificial (IA): capaz de aprender padrões e tomar decisões com base em grandes volumes de dados. Vantagem: alta escalabilidade e autonomia. Limitação: dependência de dados de qualidade e risco de viés algorítmico.

Sistemas especialistas: utilizam regras pré-definidas para tomada de decisão. Vantagem: precisão em domínios específicos. Limitação: dificuldade de adaptação a novos contextos.

Modelagem matemática: representação formal de problemas por meio de equações. Vantagem: rigor e previsibilidade. Limitação: complexidade na formulação e necessidade de conhecimento técnico.

Cada abordagem possui vantagens e limitações que devem ser consideradas conforme o contexto do problema. O pensamento computacional, por sua flexibilidade e aplicabilidade transversal, destaca-se como uma base sólida para a construção de soluções inovadoras e eficazes.


# **Pipeline da Solução**

Um pipeline eficaz de processamento de dados para problemas computacionais deve seguir uma sequência estruturada de etapas que assegurem qualidade, eficiência e escalabilidade. A seguir, apresento e discuto um modelo genérico de pipeline, flexível e adaptável a diferentes contextos de análise de dados:

Etapas do Pipeline de Processamento de Dados
Coleta de Dados

Fontes: sensores, APIs, bancos de dados, arquivos CSV, web scraping.

Ferramentas: Python (pandas, requests), Apache NiFi, Talend.

Objetivo: reunir dados brutos relevantes para o problema.

Pré-processamento

Limpeza: remoção de valores nulos, duplicados, inconsistências.

Transformação: normalização, codificação de variáveis categóricas, conversão de formatos.

Vantagem: melhora a qualidade dos dados e reduz ruído.

Análise Exploratória (EDA)

Estatísticas descritivas, visualizações (histogramas, boxplots, correlações).

Ferramentas: Jupyter Notebook, Matplotlib, Seaborn.

Objetivo: entender padrões, tendências e possíveis outliers.

Seleção de Características

Técnicas: PCA, correlação, seleção baseada em importância (e.g., Random Forest).

Vantagem: reduz dimensionalidade e melhora desempenho do modelo.

Modelagem

Algoritmos: regressão, árvores de decisão, redes neurais, SVM.

Validação cruzada e ajuste de hiperparâmetros.

Ferramentas: Scikit-learn, TensorFlow, PyTorch.

Objetivo: criar um modelo preditivo ou classificatório eficaz.

Avaliação

Métricas: acurácia, precisão, recall, F1-score, RMSE.

Vantagem: garante que o modelo seja confiável e generalizável.
